```python
from pathlib import Path

import pandas as pd

RESULTS = Path('mainnet_results')
DATASET_PREFIX = 'mainnet'
```

## Dataset Scope

```python
scope = pd.read_csv(RESULTS / 'notebook_dataset_scope.csv')
if DATASET_PREFIX:
    scope = scope[scope['name'].astype(str).str.startswith(DATASET_PREFIX)]
scope[[column for column in [
    'name',
    'flow_count',
    'candidate_flow_count',
    'passive_event_count',
    'log_event_count',
    'false_positive_event_count',
    'simulation_seed',
    'block_arrival_min_bytes',
] if column in scope.columns]]
```
